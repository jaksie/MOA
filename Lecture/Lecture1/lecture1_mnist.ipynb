```python
import numpy as np, pandas as pd, seaborn as sns
from sklearn.datasets import fetch_openml
import matplotlib.pyplot as plt
```

```python
X, y = fetch_openml('mnist_784', version=1, return_X_y=True) # loads the dataset in a certain way
```

```python
print(X)
X_norm = X.to_numpy()
```

```python
print(X_norm)
display(pd.DataFrame(X_norm))
```

```python
X = X.to_numpy() / 255.0
y = y.astype(int)

fig = plt.figure(figsize=(8, 8))
fig.subplots_adjust(
    left=0, right=1, bottom=0, top=1,
    hspace=0.05, wspace=0.05
)

for i in range(100):
    ax = fig.add_subplot(10, 10, i + 1, xticks=[], yticks=[])
    ax.imshow(X[i].reshape(28, 28),
              cmap=plt.cm.binary,
              interpolation='nearest') # sharper images than for 'bicubic' or 'bilinear'
    ax.text(0, 7, str(y[i]), color='red')

plt.show()
```
